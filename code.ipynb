# EDA

```python
import pandas as pd

# Load the dataset
file_path = "../../pew_data/2023-24 RLS Public Use File Feb 19.csv"
df = pd.read_csv(file_path)

# Display basic info about the dataset
df.info()

# Show the first few rows
df.head()
```

```python
# Get summary statistics for numerical columns
summary_stats = df.describe()

# Print summary statistics
print(summary_stats)
```

```python
# Filter the dataset to exclude rows where the value is 99
df_exclude_refused = df[df != 99]

# Calculate summary statistics for the filtered DataFrame
summary_stats = df_exclude_refused.describe()

# Show the summary statistics
print(summary_stats)
```

```python
# Calculate the mean for a specific column, excluding 99
mean_happiness = df['HAPPY'][df['HAPPY'] != 99].mean()

print(f"Mean happiness (excluding 99): {mean_happiness}")
```
