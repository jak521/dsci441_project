## Success Metric

We will use the HAPPY feature as a simple measure of success since happiness is most important in life. \
We will also combine HAPPY, INC_SDT1 (income level), and EDUCREC (highest educational attainment) for a more advanced success outcome

```python
# Create HAPPY_INVERTED where 1 and 3 are swapped
df_exclude_refused['HAPPY_INVERTED'] = df_exclude_refused['HAPPY'].map({1: 3, 2: 2, 3: 1})

# Create SUCCESS as the sum of INC_SDT1, HAPPY_INVERTED, and EDUREC
df_exclude_refused['SUCCESS'] = df_exclude_refused['INC_SDT1'] + df_exclude_refused['HAPPY_INVERTED'] + df_exclude_refused['EDUCREC']
```

We need to check that the frequency of each feature and outcome combo is at least 5 in order to run a Chi Square test

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_exclude_refused['RELPER'], df_exclude_refused['HAPPY'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Check expected frequencies
print("Expected frequencies:")
print(expected)
```

Run the Chi-Square test to test if there is a statistically significant relationship between RELPPER and both HAPPY and SUCCESS

```python
# Chi-Square Test function
def chi_square_test(df, var1, var2):
    # Create contingency table
    contingency_table = pd.crosstab(df[var1], df[var2])

    # Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # Output the results
    print(f"Chi-Square Test for {var1} vs {var2}:")
    print(f"Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
    print("Expected frequencies:")
    print(expected)
    print("\n")
    return p

# Run Chi-Square test for RELPER vs HAPPY
p_value_happy = chi_square_test(df_exclude_refused, 'RELPER', 'HAPPY')

# Run Chi-Square test for RELPER vs HAPPY_INVERTED
p_value_success = chi_square_test(df_exclude_refused, 'RELPER', 'SUCCESS')

# Conclusion
if p_value_happy < 0.05:
    print("There is a statistically significant relationship between RELPER and HAPPY.")
else:
    print("There is no statistically significant relationship between RELPER and HAPPY.")

if p_value_success < 0.05:
    print("There is a statistically significant relationship between RELPER and SUCCESS.")
else:
    print("There is no statistically significant relationship between RELPER and SUCCESS.")
```

```python
from scipy.stats import f_oneway

# Perform ANOVA to compare the means of SUCCESS across different levels of RELPER
def anova_test(df_exclude_refused, feature, outcome):
    groups = [df_exclude_refused[df_exclude_refused[feature] == level][outcome] for level in df_exclude_refused[feature].unique()]
    
    # Perform one-way ANOVA
    f_stat, p_val = f_oneway(*groups)
    
    print(f"ANOVA Test for {feature} vs {outcome}:")
    print(f"F-statistic: {f_stat}, p-value: {p_val}")
    
    return p_val

# Run ANOVA for RELPER and SUCCESS
p_value_success = anova_test(df_exclude_refused, 'RELPER', 'SUCCESS')

# Conclusion
if p_value_success < 0.05:
    print("There is a statistically significant relationship between RELPER and SUCCESS.")
else:
    print("There is no statistically significant relationship between RELPER and SUCCESS.")
```
