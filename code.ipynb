```python
import matplotlib.pyplot as plt

color_map = {1: '#99FF99',  # green
             2: '#FF9999',  # red
             3: '#66B2FF',  # blue
             99: '#BEBEBE'}  # grey

# Function to plot pie chart for a specific feature
def plot_pie_chart(data, feature_name):
    # Calculate the proportions
    value_counts = data[feature_name].value_counts()
    
    # Assign colors based on the defined color_map
    colors = [color_map.get(val, '#FFFFFF') for val in value_counts.index]
    
    # Plot the pie chart
    plt.pie(value_counts, labels=value_counts.index, colors=colors, autopct='%1.1f%%', startangle=90)
    plt.title(f'Proportion of Outcomes for {feature_name}')
    plt.show()

# Plot pie charts for CHNG_A, CHNG_B, and CHNG_C
print('CHNG Questions have outcomes of 1: Change for the better, 2: Change for the worse, 3: Indifferent, 99: Refused to Answer \n')
print('CHNG_A: Growing population of immigrants \n')
plot_pie_chart(df, 'CHNG_A')
print('CHNG_B: More women in workforce \n')
plot_pie_chart(df, 'CHNG_B')
print('CHNG_C: Greater acceptance of transgender people')
plot_pie_chart(df, 'CHNG_C')
print('DIVRELPOP: Diversity of religion and aetheism')
plot_pie_chart(df, 'DIVRELPOP')
print('DIVRACPOP: Diversity of race')
plot_pie_chart(df, 'DIVRACPOP')
```

```python
# Mapping of outcome codes to their descriptive labels
outcome_mapping = {
    1100: 'Evangelical Protestant Tradition',
    1200: 'Mainline Protestant Tradition',
    1300: 'Historically Black Protestant Tradition',
    10000: 'Catholic',
    20000: 'Church of Jesus Christ of Latter-day Saints (Mormon)',
    30000: 'Orthodox Christian',
    40001: 'Jehovah\'s Witness',
    40002: 'Other Christian',
    50000: 'Jewish',
    60000: 'Muslim',
    70000: 'Buddhist',
    80000: 'Hindu',
    90001: 'Other World Religions',
    90002: 'Other Faiths',
    100000: 'Religiously Unaffiliated',
    900000: 'Don\'t know/refused'
}

# Function to plot pie chart with labels pointing to each wedge and descriptive labels
def plot_pie_chart_with_labels_and_mapping(data, feature_name, mapping):
    # Calculate the proportions
    value_counts = data[feature_name].value_counts()

    # Map the numeric values to their descriptive labels using the mapping
    labels = [mapping.get(val, str(val)) for val in value_counts.index]

    # Plot the pie chart
    wedges, texts, autotexts = plt.pie(
        value_counts, 
        labels=labels, 
        autopct='%1.1f%%', 
        startangle=90,
        wedgeprops={'edgecolor': 'black'},  # Optional: adds black borders to wedges
        pctdistance=0.85,  # Adjust this value to control the distance of the percentage labels
        labeldistance=1.1  # Controls the distance of the feature labels
    )
    
    # Set title
    plt.title(f'Proportion of Outcomes for {feature_name}: Family Religious Tradition')
    
    # Add labels to point to each wedge
    for text in texts:
        text.set_fontsize(10)
    
    # Show the plot
    plt.show()

# Plot the pie chart for DENOMREC2 with descriptive labels
plot_pie_chart_with_labels_and_mapping(df, 'RELTRAD', outcome_mapping)
```
