```python
import numpy as np

df_exclude_refused['HAPPY_INVERTED'] = df_exclude_refused['HAPPY'].replace({1: 3, 2: 2, 3: 1})  # Swap 1 and 3, keep 2 the same

df_exclude_refused['SUCCESS'] = np.where(
    df_exclude_refused[['INC_SDT1', 'HAPPY_INVERTED', 'EDUCREC']].eq(99).any(axis=1),
    np.nan,  # Set to NaN if any value is 99
    df_exclude_refused['INC_SDT1'] + df_exclude_refused['HAPPY_INVERTED'] + df_exclude_refused['EDUCREC']
)

df_success_exclude_refused = df_exclude_refused.dropna(subset=['SUCCESS'])
```

We need to check that the frequency of each feature and outcome combo is at least 5 in order to run a Chi Square test

```python
from scipy.stats import chi2_contingency

df_relper_happy_exclude_refused = df_exclude_refused[(df_exclude_refused['RELPER'] != 99) & (df_exclude_refused['HAPPY'] != 99)]
df_relper_success_exclude_refused = df_success_exclude_refused[df_success_exclude_refused['RELPER'] != 99]

# Create a contingency table
contingency_table_relper_happy = pd.crosstab(df_relper_happy_exclude_refused['RELPER'], df_relper_happy_exclude_refused['HAPPY'])
contingency_table_relper_success = pd.crosstab(df_relper_success_exclude_refused['RELPER'], df_relper_success_exclude_refused['SUCCESS'])

# Perform the Chi-Square test
chi2, p, dof, expected_relper_happy = chi2_contingency(contingency_table_relper_happy)
chi2, p, dof, expected_relper_success = chi2_contingency(contingency_table_relper_success)

# Check expected frequencies
print("Expected RELPER / HAPPY frequencies:")
print(expected_relper_happy)
print("\n Expected RELPER / SUCCESS frequencies:")
print(expected_relper_success)
```

Run the Chi-Square test to test if there is a statistically significant relationship between RELPPER and both HAPPY and SUCCESS

```python
# Chi-Square Test function
def chi_square_test(df, var1, var2):
    # Create contingency table
    contingency_table = pd.crosstab(df[var1], df[var2])

    # Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # Output the results
    print(f"Chi-Square Test for {var1} vs {var2}:")
    print(f"Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
    print("Expected frequencies:")
    print(expected)
    print("\n")
    return p

# Run Chi-Square test for RELPER vs HAPPY
p_value_happy = chi_square_test(df_relper_happy_exclude_refused, 'RELPER', 'HAPPY')

# Run Chi-Square test for RELPER vs HAPPY_INVERTED
p_value_success = chi_square_test(df_relper_success_exclude_refused, 'RELPER', 'SUCCESS')

# Conclusion
if p_value_happy < 0.05:
    print("There is a statistically significant relationship between RELPER and HAPPY.")
else:
    print("There is no statistically significant relationship between RELPER and HAPPY.")

if p_value_success < 0.05:
    print("There is a statistically significant relationship between RELPER and SUCCESS.")
else:
    print("There is no statistically significant relationship between RELPER and SUCCESS.")
```

```python
from scipy.stats import chi2_contingency

df_usgen_relper_exclude_refused = df_exclude_refused[(df_exclude_refused['USGEN'] != 99) & (df_exclude_refused['RELPER'] != 99)]
df_usgen_success_exclude_refused = df_success_exclude_refused[df_success_exclude_refused['USGEN'] != 99]

# Create a contingency table
contingency_table_usgen_relper = pd.crosstab(df_usgen_relper_exclude_refused['USGEN'], df_usgen_relper_exclude_refused['RELPER'])
contingency_table_usgen_success = pd.crosstab(df_usgen_success_exclude_refused['USGEN'], df_usgen_success_exclude_refused['SUCCESS'])

# Perform the Chi-Square test
chi2, p, dof, expected_usgen_relper = chi2_contingency(contingency_table_usgen_relper)
chi2, p, dof, expected_usgen_success = chi2_contingency(contingency_table_usgen_success)

# Check expected frequencies
print("Expected USGEN / RELPER frequencies:")
print(expected_usgen_relper)
print("\n Expected USGEN / SUCCESS frequencies:")
print(expected_usgen_success)
```

```python
# Chi-Square Test function
def chi_square_test(df, var1, var2):
    # Create contingency table
    contingency_table = pd.crosstab(df[var1], df[var2])

    # Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # Output the results
    print(f"Chi-Square Test for {var1} vs {var2}:")
    print(f"Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
    print("Expected frequencies:")
    print(expected)
    print("\n")
    return p

# Run Chi-Square test for USGEN vs RELPER
p_value_usgen_relper = chi_square_test(df_usgen_relper_exclude_refused, 'USGEN', 'RELPER')

# Run Chi-Square test for USGEN vs SUCCESS
p_value_usgen_success = chi_square_test(df_usgen_success_exclude_refused, 'USGEN', 'SUCCESS')

# Conclusion
if p_value_usgen_relper < 0.05:
    print("There is a statistically significant relationship between USGEN and RELPER.")
else:
    print("There is no statistically significant relationship between USGEN and RELPER.")

if p_value_usgen_success < 0.05:
    print("There is a statistically significant relationship between USGEN and SUCCESS.")
else:
    print("There is no statistically significant relationship between USGEN and SUCCESS.")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt

# Filter out any RELPER == 99 (refused)
df_tree = df_relper_happy_exclude_refused[df_relper_happy_exclude_refused['USGEN'] != 99].copy()
df_tree['CURREL_SEGMENTED_ENC'] = df_tree['CURREL_SEGMENTED'].astype('category').cat.codes

# Define X and y
X = df_tree[['RELPER', 'CURREL_SEGMENTED_ENC', 'USGEN']]
y = df_tree['HAPPY_ONE_THREE']


# Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Initialize and train the Decision Tree
#tree_model = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_model = DecisionTreeClassifier( random_state=42)

tree_model.fit(X_train, y_train)

# Predict and evaluate
y_pred = tree_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Visualize the tree
plt.figure(figsize=(10, 6))
plot_tree(tree_model, feature_names=['RELPER', 'CURREL_SEGMENTED_ENC', 'USGEN'], class_names=['Not Happy', 'Very Happy'], filled=True)
plt.title("Decision Tree: RELPER predicting HAPPY_ONE_THREE")
plt.show()
```
